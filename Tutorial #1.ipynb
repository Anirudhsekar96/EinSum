```python
import numpy as np
from functools import partial
import timeit
%matplotlib inline

import matplotlib.pyplot as plt

from tqdm.auto import tqdm, trange
```

```python
ein_time_list = list()
np_time_list = list()

def ein(mat1,mat2,mat3):
    '''
        Returns the einsum variant of the problem to be solved
    '''
    np.einsum('ik -> ', mat1)

def np_(mat1,mat2,mat3):
    """
        Returns the np standard variant of the problem to be solved
    """
    np.sum(mat1)
    
for k in trange(200):
    
    mat1 = np.random.random((k,k))
    mat2 = np.random.random((k,k))
    mat3 = np.random.random((k,k))
    
    ein_par = partial(ein,mat1,mat2,mat3)
    np_par = partial(np_,mat1,mat2,mat3)
    ein_time_list.append(timeit.timeit(stmt=ein_par, number=1000))
    np_time_list.append(timeit.timeit(stmt=np_par, number=1000))

plt.figure(figsize=(10,5))
plt.plot(ein_time_list)
plt.plot(np_time_list)
plt.legend(['EIN','NP'])
plt.title('Comparison of running times')
plt.show()
```
